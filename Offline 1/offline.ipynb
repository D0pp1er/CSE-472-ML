# Data Cleaning

2.1 Replace missing values with the column-wise mean

```python

df[df.isnull().any(axis=1)]
```

```python
df[df.duplicated()]
```

```python
df.dtypes
```

```python
object_dtypes = df.select_dtypes(include='object').columns
print(f"Object columns: {object_dtypes}")
```

```python
for column in object_dtypes:
    df[column] = df[column].astype('category')

df.dtypes
```

```python
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['category']).columns

df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())

df[categorical_cols] = df[categorical_cols].apply(lambda x: x.fillna(x.mode()[0]))

df.isnull().sum().sum() 
df.isnull().sum()
df.dtypes
```

```python
df.drop_duplicates(inplace=True)
df.dtypes
```

```python
df.duplicated().sum()
```

```python
features = df.drop(columns=['Attrition'])
target = df['Attrition']
```

```python
from sklearn.preprocessing import LabelEncoder

target
label_encoder = LabelEncoder()
target_encoded = label_encoder.fit_transform(target)

for col in features.select_dtypes(include=['category']).columns:
    if features[col].nunique() == 2:
        features[col] = label_encoder.fit_transform(features[col])



features = pd.get_dummies(features)

```

```python
target_encoded
```

```python
# Count the data types in the features DataFrame
dtype_counts = features.dtypes.value_counts()
print("Data type counts in features:")
print(dtype_counts)

# Print the columns with only 1 unique value
for col in features.columns:
    unique_values = features[col].unique()
    if len(unique_values) == 1:
        print(f"Column: {col}, Unique Values: {unique_values}")
count = 0
# Print columns with more than 2 unique values or exactly 2 unique values but not {0, 1}
for col in features.columns:
    unique_values = features[col].unique()
    if len(unique_values) > 2 or (len(unique_values) == 2 and set(unique_values) != {0, 1}):
        # print(f"Column: {col}, Unique Values: {unique_values}")
        count+=1

print(f"Columns with more than 2 unique values or exactly 2 unique values but not {0, 1}: {count}")
```

```python
def scale_features(features, scaling_method):
    from sklearn.preprocessing import StandardScaler, MinMaxScaler
    if scaling_method == 'standard':
        scaler = StandardScaler()
    elif scaling_method == 'minmax':
        scaler = MinMaxScaler()

    for col in features.columns:
        unique_values = features[col].unique()
        if len(unique_values) > 2 or (len(unique_values) == 2 and set(unique_values) != {0, 1}):
            features[col] = scaler.fit_transform(features[[col]])
    
    # columns_to_scale = features[cols_with_more_than_2_unique_values].columns
    # features[cols_with_more_than_2_unique_values] = scaler.fit_transform(features[cols_with_more_than_2_unique_values])
    
    
    
    
    
    
```

```python
scale_features(features, 'standard')
# scale_features(features, 'minmax')
```

```python
# import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'features' is your DataFrame with numeric features and 'target' is your target column
# Calculate the correlation of each feature with the target column
features_df = pd.DataFrame(features, columns=features.columns)
target_df = pd.DataFrame(target_encoded,columns=['Attrition'])
target_series = target_df['Attrition']  
correlation = features_df.corrwith(target_series)
# Display the correlation values
print("Correlation of each feature with the target column:")
print(correlation)


```

```python
# Create a correlation matrix
correlation_matrix = features_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()

```

```python
# Select the top 20 features with the highest correlation with the target variable
top_20_features = correlation.abs().sort_values(ascending=False).head(20).index
print("Top 20 features with the highest correlation with the target variable:")
print(top_20_features)
final_features = features_df[top_20_features]

```

```python
correlation_matrix = final_features.corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Top 20 Features')
plt.show()
plt.savefig('correlation_heatmap.png')  
```

```python
# Create 1D scatter plots for the top 20 features 
class_0 = features_df.loc[target_df['Attrition'] == 0]
class_1 = features_df.loc[target_df['Attrition'] == 1]
import numpy as np
for feature in top_20_features:
    # plt.figure(figsize=(8, 6))
    # plt.scatter(features[feature][class_0], [0] * len(features[feature][class_0]), color='b', label='0', alpha=0.5)
    # plt.scatter(features[feature][class_1], [0] * len(features[feature][class_1]), color='r', label='1', alpha=0.5)
    plt.plot(class_0[feature],np.zeros_like(class_0[feature]), 'o', label='0')
    plt.plot(class_1[feature],np.zeros_like(class_1[feature]), 'o', label='1')


    plt.title(f"1D Scatter Plot of Feature: {feature}")
    plt.legend(['0', '1'])
    plt.savefig(f"Figures/{feature}_scatter_plot.png")
    plt.close()

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(final_features, target_encoded, test_size=0.2, random_state=42)
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy*100:.2f}%")
```
